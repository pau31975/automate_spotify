```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
json_list_endsong = glob.glob('data/endsong*.json')
json_list_SH = glob.glob('data/StreamingHistory*.json')

df_endsong = pd.DataFrame()
df_SH = pd.DataFrame()

for i in json_list_endsong:
    df_endsong = pd.concat([df_endsong, pd.read_json(i)], ignore_index=True)
    
for i in json_list_SH:
    df_SH = pd.concat([df_SH, pd.read_json(i)], ignore_index=True)
```

```python
# convert to datetime
df_SH['endTime'] = pd.to_datetime(df_SH['endTime'], format='%Y-%m-%d %H:%M')
df_endsong['ts'] = pd.to_datetime(df_endsong['ts'], format='%Y-%m-%dT%H:%M:%SZ')
```
